# Image preprocessing for model training

```python
image_width = 640 # we must maitain tha same proportions of the images 
image_height = 640
image_color_channel = 3 # 3 color dimensions RGB
image_color_channel_size = 255
image_size = (image_width, image_height)
img_shape = image_size + (image_color_channel,)

batch_size = 32 # number of features that I will bring at a time from my dataset
epochs = 30 # number of times that I will iterate trought my dataset
learning_rate = 0.0001

class_names = ['dry', 'normal', 'oily']
```

```python
train = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split = 0.2,
    subset = 'training',
    seed = 568,
    image_size = (image_height,image_width),
    batch_size = batch_size
)
```

```python
test = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    seed = 568,
    image_size = (image_height,image_width),
    batch_size = batch_size
)
```

```python
validation = tf.keras.utils.image_dataset_from_directory(
    train_dir,
     validation_split = 0.2,
    subset = 'validation',
    seed = 568,
    image_size = (image_height,image_width),
    batch_size = batch_size
)
```

# Creating a Machine Learning Supervised Learning model - type Image Classification

```python
model = tf.keras.models.Sequential([
    
    # Input:
    tf.keras.layers.Input(shape=(640, 640,  3)),

    # Reescaling Layers
    tf.keras.layers.Rescaling(1./255),
    
    # Padding Layers
    tf.keras.layers.Flatten(),
    
    # Hidden Layers:
    tf.keras.layers.Dense(128,activation=tf.nn.relu),

    # Output:
    tf.keras.layers.Dense(3, activation=tf.nn.softmax) # Softmax activation is used for cathegorical data and 3 are the dimensions in our dataset
])
```

```python
model.summary()
```

## Training

Compiling

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
epocas = 10

history = model.fit(
    train,
    validation_data = validation,
    epochs = epocas # qtd de iterações q o algoritmo irá fazer
)
```
